```python
from string import ascii_lowercase
from string import ascii_uppercase
import re

cleanr = re.compile('<.*?>')
def cleanhtml(raw_html):
    cleantext = re.sub(cleanr, '', raw_html)
    return cleantext

def lstrip(s):
    s = cleanhtml(s)
    for i in range(len(s)):
        if s[i] in ascii_lowercase or s[i] in ascii_uppercase:
            return s[i:]
        else:
            continue
    return ''
```

```python
from string import punctuation
from string import digits
punctuation = punctuation + ''
real = []
corpus = ''
for person in tqdm(all_results):
    profile = person.get('profile')
    if profile is None: continue
    school = profile.get('school')
    experience = profile.get('experience')
    if experience is None or len(experience) == 0: continue
    for job in experience:
        desc = job.get('description')
        desc = desc.replace('</p>', '').replace('<p>', '')
        lines = desc.strip().split('\n')
        for line in lines:
            cleaned = line.lstrip(punctuation).lstrip(digits).strip()
            cleaned = lstrip(cleaned)
            if len(cleaned.split(' ')) < 4 or len(cleaned.split(' ')) > 24: continue
            real.append(cleaned.lower())
            corpus += cleaned + ' '
```
